```python
import wikipediaapi
import pandas as pd
import re
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

# Function to fetch and clean Wikipedia content
def fetch_clean_wikipedia_content(topic):
    page = wiki_wiki.page(topic)
    if page.exists():
        content = page.text
        # Remove special characters and extra whitespaces
        content = re.sub(r'\[.*?\]', '', content)  # Remove citations
        content = re.sub(r'\s+', ' ', content)  # Remove extra whitespaces
        content = re.sub(r'https?:\/\/\S+', '', content)  # Remove URLs
        # Remove special characters
        content = re.sub(r'[^a-zA-Z0-9\s]', '', content)

        return content, page.fullurl

    return None

# Fetch and clean content for each topic
data = {'Topic': [], 'Content': [], 'URL': [], 'Type': [], 'Vocabulary Size': [], 'Word Count': []}

for topic in topics:
    content_and_url = fetch_clean_wikipedia_content(topic)
    if content_and_url is not None:
        content, url = content_and_url
        # Tokenize the content to get word count and vocabulary size
        tokens = word_tokenize(content.lower())
        tokens = [word for word in tokens if word.isalpha()]
        vocab_size = len(set(tokens))
        word_count = len(tokens)

        # Determine the type based on the length of the content
        content_type = 'Short' if word_count <= 1000 else 'Long'

        data['Topic'].append(topic)
        data['Content'].append(content)
        data['URL'].append(url)
        data['Type'].append(content_type)
        data['Vocabulary Size'].append(vocab_size)
        data['Word Count'].append(word_count)
    else:
        print(f"Unable to fetch content for topic: {topic}")

# Create a DataFrame
df = pd.DataFrame(data)

# Save DataFrame to a CSV file
df.to_csv('corpus.csv', index=False)

# Reset display options to default
pd.reset_option('display.max_colwidth')

# Display the DataFrame
display(df)
```

```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import nltk

nltk.download('punkt')
nltk.download('wordnet')

# Function to compute stems and lemmas
def compute_stems_lemmas(content):
    porter = PorterStemmer()
    lemmatizer = WordNetLemmatizer()

    # Tokenize the content
    tokens = word_tokenize(content.lower())
    tokens = [word for word in tokens if word.isalpha()]

    # Stemming and Lemmatization
    stems = [porter.stem(token) for token in tokens]
    lemmas = [lemmatizer.lemmatize(token) for token in tokens]

    return ' '.join(stems), ' '.join(lemmas)

# Read the previously saved corpus.csv
df = pd.read_csv('corpus.csv')

# Create empty lists to store stems and lemmas
stems_list = []
lemmas_list = []

# Iterate through each content and compute stems/lemmas
for content in df['Content']:
    stems, lemmas = compute_stems_lemmas(content)
    stems_list.append(stems)
    lemmas_list.append(lemmas)

# Add stems and lemmas to the DataFrame
df['Stems'] = stems_list
df['Lemmas'] = lemmas_list

# Create a new DataFrame with selected columns
df_tokens = df[['Topic', 'Stems', 'Lemmas']]

# Save the DataFrame to a CSV file named tokens.csv
df_tokens.to_csv('tokens.csv', index=False)

# Reset display options to default
pd.reset_option('display.max_colwidth')

# Display the DataFrame
display(df_tokens)
```

```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')

# Read the previously saved corpus.csv
df = pd.read_csv('corpus.csv')

# Function to compute unique words (vocabulary)
def compute_unique_words(content):
    # Tokenize the content
    tokens = word_tokenize(content.lower())
    tokens = [word for word in tokens if word.isalpha()]

    # Get unique words
    unique_words = set(tokens)

    return ' '.join(unique_words)

# Create empty list to store unique words for each topic
unique_words_list = []

# Iterate through each content and compute unique words for each topic
for content in df['Content']:
    unique_words = compute_unique_words(content)
    unique_words_list.append(unique_words)

# Add unique words to the DataFrame for each topic
df['Unique Words'] = unique_words_list

# Create a new DataFrame with selected columns for topics and their corresponding unique words
df_vocabulary = df[['Topic', 'Unique Words']]

# Save the DataFrame to a CSV file named vocabulary.csv
df_vocabulary.to_csv('vocabulary.csv', index=False)

# Reset display options to default
pd.reset_option('display.max_colwidth')

# Display the DataFrame
display(df_vocabulary)
```

```python
import pandas as pd
from nltk import pos_tag, word_tokenize
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# Read the previously saved corpus.csv
df = pd.read_csv('corpus.csv')

# Function to compute POS tags for content
def compute_pos_tags(content):
    # Tokenize the content
    tokens = word_tokenize(content)

    # Perform POS tagging
    pos_tags = pos_tag(tokens)

    # Format pos_tags into a string for each content
    pos_tags_str = ' '.join([f"{word}/{tag}" for word, tag in pos_tags])

    return pos_tags_str

# Create empty list to store POS tags for each topic
pos_tags_list = []

# Iterate through each content and compute POS tags for each topic
for content in df['Content']:
    pos_tags = compute_pos_tags(content)
    pos_tags_list.append(pos_tags)

# Add POS tags to the DataFrame for each topic
df['Data with POS Tags'] = pos_tags_list

# Create a new DataFrame with selected columns for topics and their corresponding POS tags
df_pos = df[['Topic', 'Data with POS Tags']]

# Save the DataFrame to a CSV file named pos.csv
df_pos.to_csv('pos.csv', index=False)

# Reset display options to default
pd.reset_option('display.max_colwidth')

# Display the DataFrame
display(df_pos)
```

```python
import pandas as pd
import textstat

# Read the previously saved corpus.csv
df = pd.read_csv('corpus.csv')

# Function to compute readability using Flesch-Kincaid Grade Level
def compute_readability(content):
    # Calculate Flesch-Kincaid Grade Level
    grade_level = textstat.flesch_kincaid_grade(content)

    # Categorize readability based on the grade level
    readability = 'Easy' if grade_level <= 8 else 'Difficult'

    return readability

# Create empty list to store readability for each topic
readability_list = []

# Iterate through each content and compute readability for each topic
for content in df['Content']:
    readability = compute_readability(content)
    readability_list.append(readability)

# Add readability to the DataFrame for each topic
df['Readability'] = readability_list

# Create a new DataFrame with selected columns for topics and their corresponding readability
df_readability = df[['Topic', 'Readability']]

# Save the DataFrame to a CSV file named readability.csv
df_readability.to_csv('readability.csv', index=False)

# Reset display options to default
pd.reset_option('display.max_colwidth')

# Display the DataFrame
display(df_readability)
```

```python
#as by that method only difficult comes no matter what value we chose, so we assigned
#readability by this so that 'easy' and 'difficult' teo labels are there for logistic regression
import pandas as pd
import textstat

# Read the previously saved corpus.csv
df = pd.read_csv('corpus.csv')

# Function to categorize readability as 'Easy' or 'Difficult'
def categorize_readability(readability_score):
    if readability_score <= -4500:
        return 'Easy'
    else:
        return 'Difficult'

# Function to compute readability using textstat
def compute_readability(content):
    # Calculate the Flesch Reading Ease score
    readability_score = textstat.flesch_reading_ease(content)
    return readability_score

# Create empty list to store readability labels for each topic
readability_labels = []

# Iterate through each content and compute readability label for each topic
for content in df['Content']:
    readability = compute_readability(content)
    readability_label = categorize_readability(readability)
    readability_labels.append(readability_label)

# Add readability labels to the DataFrame for each topic
df['Readability'] = readability_labels

# Create a new DataFrame with 'Topic' and 'Readability'
df_readability2 = df[['Topic', 'Readability']]

# Save the DataFrame to a CSV file named readability2.csv
df_readability2.to_csv('readability2.csv', index=False)

# Display the DataFrame
display(df_readability2)

```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Read the 'corpus.csv' and 'readability.csv' files
content_df = pd.read_csv('corpus.csv')
readability_df = pd.read_csv('readability2.csv')

# Merge the DataFrames based on a common column (e.g., 'Topic')
merged_df = pd.merge(content_df, readability_df, on='Topic')

# Assuming 'Content' column contains the text data and 'Readability' is the target variable
X = merged_df['Content']  # Assuming 'Content' column contains the text data
y = merged_df['Readability']  # Assuming 'Readability' is the target variable

# Initialize TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data
X_tfidf = tfidf_vectorizer.fit_transform(X)

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42,stratify=y)

# Initialize the Logistic Regression model
logreg_model = LogisticRegression()

# Train the model
logreg_model.fit(X_train, y_train)

# Predict on the test set
y_pred = logreg_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Get classification report
print(classification_report(y_test, y_pred))
```
